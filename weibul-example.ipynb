```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Define the scale parameter (η)
eta = 25

# Define the shape parameters (β)
beta_values = [1, 2, 3, 5]

# Define the time range for the plot
time = np.linspace(0, 50, 500)

# Plot Weibull PDFs for different β values
plt.figure(figsize=(10, 6))

for beta in beta_values:
    pdf = weibull_min.pdf(time, beta, scale=eta)
    plt.plot(time, pdf, label=f'β = {beta}')

# Mark the scale parameter
plt.axvline(x=eta, color='r', linestyle='--', label=f'Lifetime equals scale ({eta} years)')

# Add labels and legend
plt.title(f'Weibull Probability Density Function (PDF) for Different Shape Parameters (β), η={eta} years')
plt.xlabel('Time (years)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
for beta in beta_values:
    cdf = weibull_min.cdf(time, beta, scale=eta)
    plt.plot(time, cdf, label=f'β = {beta}')

# Mark the economic lifetime at 20 years
plt.axvline(x=20, color='r', linestyle='--', label='Economic Lifetime (20 years)')

# Add labels and legend
plt.title(f'Weibull Cumulative Distribution Function (CDF) for Different Shape Parameters (β), η={eta} years')
plt.xlabel('Time (years)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python

# Weibull parameters for different quality categories
params = {
    'Category A': {'beta': 4.5, 'eta': 18},
    'Category B': {'beta': 3.0, 'eta': 12},
    'Budget': {'beta': 2.0, 'eta': 9}
}

# Time range for the plot
time = np.linspace(0, 25, 400)

# Plot Weibull CDFs for each category
plt.figure(figsize=(10, 6))
for category, param in params.items():
    beta = param['beta']
    eta = param['eta']
    cdf = weibull_min.cdf(time, beta, scale=eta)
    plt.plot(time, cdf, label=f'{category} (β={beta}, η={eta} years)')

# Mark the economic lifetime at 12 years
plt.axvline(x=12, color='red', linestyle='--', label='Economic Lifetime (12 years)')

# Add labels and legend
plt.title('Weibull CDF Comparison of Different Quality Categories')
plt.xlabel('Time (years)')
plt.ylabel('Cumulative Probability of Failure')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python
plt.figure(figsize=(10, 6))
for category, param in params.items():
    beta = param['beta']
    eta = param['eta']
    pdf = weibull_min.pdf(time, beta, scale=eta)
    plt.plot(time, pdf, label=f'{category} (β={beta}, η={eta} years)')

# Mark the economic lifetime at 12 years
plt.axvline(x=12, color='red', linestyle='--', label='Economic Lifetime (12 years)')

# Add labels and legend
plt.title('Weibull PDF Comparison of Different Quality Categories')
plt.xlabel('Time (years)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the alpha and beta parameters
params = [
    (0.5, 0.5),
    (1, 1),
    (2, 2),
    (2, 5),
    (5, 1)
]

# Generate x values
x = np.linspace(0, 1, 100)

# Plot the Beta distributions for different alpha and beta values
plt.figure(figsize=(10, 6))

for a, b in params:
    y = beta.pdf(x, a, b)
    plt.plot(x, y, label=f'alpha={a}, beta={b}')

plt.title('Beta Distribution for Different alpha and beta Values')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

```
